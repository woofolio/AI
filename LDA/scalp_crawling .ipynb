```python
import urllib.request
from bs4 import BeautifulSoup
import requests  # requests 라이브러리 추가
import time
```

```python
title_lst = []
```

```python
for page in range(1, 301):
    # 두피를 검색하였을 때 나온 질문들이 페이지 당 10개씩 있다.
    # 10페이지의 제목을 크롤링하여 100개의 텍스트를 생성하였다.
    url = 'https://kin.naver.com/search/list.naver?query=%EB%91%90%ED%94%BC&page=' + str(page)
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')

    # 한 페이지의 제목 글 저장
    title = soup.find_all(class_='_nclicks:kin.txt _searchListTitleAnchor')

    for i in title:
        title_lst.append(i.attrs['href'])  # href 속성에 있는 모든 값을 문자열로 리턴해준다.

    time.sleep(3)

print(len(title_lst))

```

```python
txt_lst = []
```

```python
for title_txt in title_lst:
    try:
        html = requests.get(title_txt).content
        soup = BeautifulSoup(html, 'html.parser')
        txt = soup.find(class_='c-heading__content').get_text()
        text = str(txt)
        text = text.replace('\n', '')  # 변수 이름 수정 (title_txt -> text)
        text = text.replace('\t', '')  # 변수 이름 수정 (title_txt -> text)
        txt_lst.append(text)

    except:
        txt = soup.find(class_='title').get_text()
        text = str(txt)
        text = text.replace('\n', '')  # 변수 이름 수정 (title_txt -> text)
        text = text.replace('\t', '')  # 변수 이름 수정 (title_txt -> text)
        txt_lst.append(text)

print(txt_lst)
```

```python
import pandas as pd

df = pd.DataFrame({'Q' : txt_lst})
df.to_excel('/content/drive/MyDrive/Colab Notebooks/scalp/scalp_project/두피관리.xlsx', index=False)
```
