# Data 전처리

```python
# List로 변환

data_ = data.text.values.tolist()

data_[:5]
```

```python
# 정규표현식을 사용해 한글 및 숫자만 추출
import re

regex = []

for i in range(len(data)):
  text = re.sub('[^0-9ㄱ-힣]', ' ', str(data_[i]))
  text = re.sub('\+', ' ', text)
  regex.append(text)
```

```python
regex[:5]
```

```python
# 명사 추출하기 위해 konlpy 설치가 필요
!pip install konlpy
```

```python
# 형태소 분석기 Okt 불러오기
from konlpy.tag import Okt

okt = Okt()
```

```python
okt.nouns(regex[0])
```

```python
# okt를 이용해 명사추출
# 약 10분 소요

noun = []

for j in range(len(regex)):
  noun.append(okt.nouns(regex[j]))
  print(str(j) + "번째 수행중")
```

```python
# Worldcloud 할 때는 리스트 안에 리스트 구조 [[]] 이것을 하나의 리스트로 만들어 주었음
# 하지만 LDA의 받는 인자는 [[a, b, c],[d, e, f]]로 구성 --> 문서별로 단어를 구분하기 위함

# nouns = sum(noun, [])

# nouns[:5]

noun[:2]
```

# 빈도추출

```python
# collections를 통해 빈도추출
from collections import Counter
```

```python
# 리스트 안에 리스트 구조 [[]] 이것을 하나의 리스트로 만들어 줘야 함
nouns = sum(noun, [])

nouns[:5]
```

```python
# 빈도 수 체크 함수
count = Counter(nouns)
```

```python
# 가장 자주 나오는 100개 단어 표출
noun_list = count.most_common(100)
```

```python
noun_list
```

# WordCloud 구현

```python
# Python에서는 wordcloud 모듈을 통해 손쉽게 wordcloud 만들기 가능
# wordcloud 설치

!pip install wordcloud
```

```python
# Wordcloud와 그것을 표출할 matplotlib 호출

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# WordCloud Setting 설정

wc = WordCloud(font_path='/content/drive/MyDrive/Colab Notebooks/dupi_project/NanumBarunGothicLight.ttf',
               background_color='white',
               width = 1000,
               height = 1000,
               max_words=100,
               max_font_size=300)
```

```python
# 빈도추출에서 만들어 둔 noun_list를 dictionary로 변환한 뒤 generate

dictionary =dict(noun_list)
wordcloud = wc.generate_from_frequencies(dictionary)

# 만약 빼고 싶은 단어(불용어)가 있다면

# dictionary = dict(noun_list)
# del(dictionary['두피'])
# wordcloud = wc.generate_from_frequencies(dictionary)
```

```python
# matplotlib를 통해 표출

plt.figure(figsize=(10, 8))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```
